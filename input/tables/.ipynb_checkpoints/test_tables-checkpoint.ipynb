```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.interpolate as interp
import asdf
import time

from copy import deepcopy

import commah
import calculate_tables as ct
import halo.tools as tools
```

# Load the default cosmology

```python
rho_crit = 2.7763458 * (10.0**11.0)
cosmo = {'sigma_8': 0.821,
         'h': 0.7,
         'omegab': 0.0463,
         'omegac': 0.233,
         'omegam': 0.0463 + 0.233,
         'omegav': 0.7207,
         'n': 0.972}
cosmo_cp = ct.convert_cosmo_commah(cosmo)
```

# Test interpolators

First we check how fast the interpolator is on a grid in (z, m200c)
with fixed cosmology.

```python
af = asdf.open("halo/input/tables/c_correa_200c.asdf")
results_c = af.tree
sigma_8 = results_c["sigma_8"]
omegam  = results_c["omegam"]
omegav  = results_c["omegav"]
n = results_c["n"]
h = results_c["h"]
c = results_c["c"]
z_range = results_c["z_range"][:]
m_range = results_c["m_range"][:]
```

```python
c_interp = interp.RegularGridInterpolator((z_range, np.log10(m_range)), c)
```

```python
t1 = time.time()
c_test = c_interp(np.vstack([np.zeros(100), np.linspace(11,15,100)]).T)
t2 = time.time()
print t2 - t1
```

Now we test how much the added cosmology interpolation slows our
process down.

```python
af_cosmo = asdf.open("c_correa_200c_cosmo.asdf")
results_c_cosmo = af_cosmo.tree
sigma_8_cosmo = results_c_cosmo["sigma_8"][:]
omegam_cosmo  = results_c_cosmo["omegam"][:]
omegav_cosmo  = results_c_cosmo["omegav"][:]
z_range_cosmo = results_c_cosmo["z_range"][:]
m_range_cosmo = results_c_cosmo["m_range"][:]
n_cosmo = results_c_cosmo["n"][:]
h_cosmo = results_c_cosmo["h"][:]
c_cosmo = results_c_cosmo["c"][:]
```

```python
c_interp_cosmo = interp.RegularGridInterpolator((sigma_8_cosmo, omegam_cosmo, omegav_cosmo, n_cosmo, h_cosmo,
                                                 z_range_cosmo, np.log10(m_range_cosmo)), c_cosmo)
```

```python
t1_cosmo = time.time()
c_test_cosmo = c_interp_cosmo(np.vstack([np.ones(100) * cosmo['sigma_8'],
                                         np.ones(100) * cosmo['omegam'],
                                         np.ones(100) * cosmo['omegav'],
                                         np.ones(100) * cosmo['n'],
                                         np.ones(100) * cosmo['h'],
                                         np.ones(100),
                                         np.linspace(11,15,100)]).T)
t2_cosmo = time.time()
print t2_cosmo - t1_cosmo
```

# Test accuracy of interpolator

Now we test how close the interpolator can get to an intermediate
value.

```python
z_test = 4
x_in = np.vstack([z_test * np.ones(50), np.linspace(5,17,50)]).T
```

```python
plt.clf()
plt.plot(np.power(10, x_in[:,1]), c_interp(x_in), label='interp')
plt.plot(m_range, commah.run(cosmology=cosmo_cp, Mi=m_range/cosmo['h'], z=z_test, mah=False)['c'].T[0],
         label='exact')
plt.loglog()
plt.legend()
plt.show()
```
